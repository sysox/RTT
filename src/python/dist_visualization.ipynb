# Binomial vs normal
### Probability distribution 

```python
n = 98 # number of trials
p = 0.4 # probability of success
mean = n*p 
std = math.sqrt(n*p*(1-p)) # standard deviation
B = binomial_data(n=n, p=p)
N = normal_data(x=B['outcomes'], mean=mean, std=std)

obs = 50
rtail_x = N['outcomes'][obs:]
rtail_pdf = N['pdf'][obs:]
```

### P-value distribution 
p-values correspond to outcomes (`x = B['outcomes']`) used in cell above !

```python
obs = 56
plt.figure(figsize=(20,6))

colors = {"B":'g','N': 'r', 'R':'b'}

x = B['outcomes']
num_points = len(x) 
plt.bar(B['outcomes'], B['pmf'], width= 0.1, color = colors['B']) 
plt.plot(N['outcomes'], N['pdf'], color = colors['N']) 
plt.fill_between(rtail_x, rtail_pdf, color = colors['R'])
plt.show()
```

```python
plt.figure(figsize=(20,6))
marker_size = 10
m = 0.004
# Binomial
plt.plot(B['less'], B['pmf'], '.',  color = colors['B'], markersize=marker_size)
plt.plot(B['less'], [-m]*num_points, '.',  color = colors['B'], markersize=marker_size)

# normal approx
plt.plot(N['less'], N['pdf'], '.',  color = colors['N'], markersize=marker_size)
plt.plot(N['less'], [-2*m]*num_points, '.',  color = colors['N'], markersize=marker_size)

plt.show()
```
